```python
#Time Resampling
```

```python
df_tesla.resample(rule = 'A').min()
```

```python
df_tesla.resample(rule = 'A').max()
```

```python
df_tesla.resample(rule = 'A').max()['Open'].plot(linestyle = '--', color = 'green')
```

```python
#QS defines quaterly start frequency
df_tesla.resample(rule = 'QS').max()['Open'].plot()
```

```python
#BA defines Business End frequency
df_tesla.resample(rule = 'BA').max()
```

```python
#BA defines Business End frequency
df_tesla.resample(rule = 'BA').max()['Open'].plot()
```

```python
#BQS defines Business quarters frequency
df_tesla.resample(rule = 'BQS').max()['Open'].plot()
```

```python
#BA defines Business End frequency
df_tesla.resample(rule = 'BA').max()['Open'].plot(kind = 'bar')
```
